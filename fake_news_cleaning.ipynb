#### Getting the rows with urls in any of the columns 

```python
import re
URLS = r'(https://)'
true_copy = true_df
fake_copy = fake_df
true_copy['urlcount'] = true_df.text.apply(lambda x: re.findall(URLS, x)).str.len()
fake_copy['urlcount'] = fake_df.text.apply(lambda x: re.findall(URLS, x)).str.len()
```

```python
true_copy['urlcount'].value_counts()
```

```python
fake_copy.loc[fake_copy['urlcount'] > 0].value_counts().sum()
```
